```python
import netCDF4 as nc
import pandas as pd
import numpy as np
import xarray as xr
import scipy
import os
import glob
from scipy import interpolate
```

```python
def getEmissions1(months, yyyy, lat1, lat2, lon1, lon2):
    end_ye = 2022
    while yyyy <= end_ye:
        GFED4 = "C:/Akmal/Telyu/Telyu/TA/Data/GFED4.1s_" + str(yyyy) + ".hdf5"
        print(GFED4)
        dataset = nc.Dataset(GFED4)
        dataframe_collection_list = []
        day, month = '01', 1
        for mm in months:
            #time = str(day)+"/"+str(month)+"/"+str(yyyy)
            time = str(yyyy)+"-"+str(mm)+"-"+str(day)+" 12:00:00"
            # Annotation: "df" is short for "data frame".
            df = pd.DataFrame()
            # Reads the latitudes and longitudes and converts them to a numpy.ndarray.
            df['longitudes']          = dataset['/lon'][:].flatten()
            df['latitudes']           = dataset['/lat'][:].flatten()
            df['time']                = time
            df['emissions']           = dataset['emissions/'+mm+'/C'][:].flatten()
            # Emissions masked for a "square" of coordinates.
            latitude_mask    = (lat1 <= df["latitudes"]) & (df["latitudes"] <= lat2)
            longitude_mask   = (lon1 <= df["longitudes"]) & (df["longitudes"] <= lon2)
            #latitude_mask    = np.sort(latitude_mask) 
            masked_df        = (df[latitude_mask & longitude_mask])
            dataframe_collection_list.append(masked_df)
            
            #print(masked_df['emissions'].max())
            #print(dataframe_collection_list)
            #print(masked_df['latitudes'].min(),'\n',masked_df['latitudes'].max())
            #print(masked_df['longitudes'].min(),'\n',masked_df['longitudes'].max())
        csvfiles = pd.concat([
                dataframe_collection_list[0],dataframe_collection_list[1],dataframe_collection_list[2],dataframe_collection_list[3],
                dataframe_collection_list[4],dataframe_collection_list[5],dataframe_collection_list[6],dataframe_collection_list[7],
                dataframe_collection_list[8],dataframe_collection_list[9],dataframe_collection_list[10],dataframe_collection_list[11]
        ])
        csvfiles.to_csv('C:/Akmal/Telyu/Telyu/TA/DatasetCSV/'+'GFED_'+str(yyyy)+'.csv', index=False, encoding='utf-8')
        print('done csv file: ',yyyy)
        yyyy=yyyy+1

    # setting the path for joining multiple files
    files = os.path.join("C:/Akmal/Telyu/Telyu/TA/DatasetCSV/", "GFED_*.csv")
    files = glob.glob(files)
    print("Mergered")
    # joining files with concat and read_csv 
    df = pd.concat(map(pd.read_csv, files), ignore_index=True)
    df.to_csv('C:/Akmal/Telyu/Telyu/TA/DatasetCSV/All/GFED_Original_98-22.csv', index=False, encoding='utf-8') 
```

```python
yyyy = 1998
months = ['01','02','03','04','05','06','07','08','09','10','11','12']
lat1, lat2, lon1, lon2 = -4.25,6.25, 108.5, 119.25
getEmissions1(months, yyyy, lat1, lat2, lon1, lon2)
```

```python
def filterGrid(lat0,lat1,lon0,lon1,dataframe,data):
    if dataframe.empty :
        data = dataframe.query("(longitude >=" +str(lon0)+ "& longitude <=" +str(lon1)+") & (latitude >=" +str(lat0)+ "& latitude <="+str(lat1)+")")
    else:
        data = pd.concat([data,dataframe.query("(longitude >=" +str(lon0)+ "& longitude <=" +str(lon1)+") & (latitude >=" +str(lat0)+ "& latitude <="+str(lat1)+")")])
    return data
```
